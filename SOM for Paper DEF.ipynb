```python
from minisom import MiniSom
import pandas as pd
import numpy as np
import json
import matplotlib.pyplot as plt
```

```python
prefix = 'Band'
numbers = range(1,32,1)
names = [f"{prefix}{n}" for n in numbers]
b_f = ['Buff', 'Frequency', 'Freq_Label']
names = names+b_f
```

```python
with open ('Data Pre Processing/JSON/nopeaks.json', 'r') as archivo:
    # Leer el contenido del archivo JSON y convertirlo en un objeto de Python
    nopeaks = json.load(archivo)

nopeaks_results = []
for file in nopeaks:
    
    resultado = list(file['resultado'])
    resultado.append(file['Buff']) #esto es para q buff quede dentro de resultado
    resultado.append(file['Frequency'])
    resultado.append(file['Freq_Label'])
    
    #esto es para q Buff y Freq queden fuera de resultados
    #resultado = {"resultado": resultado, "Buff": file['Buff'], "Frequency": file['Frequency']}
    #print(resultado)
    
    nopeaks_results.append(resultado)

data_nopeaks = np.array(nopeaks_results)
data_nopeaks_names = pd.DataFrame(data_nopeaks, columns = names)

target1 = data_nopeaks_names['Buff'].values
target1 = np.array(target1, dtype = int)
target2 = data_nopeaks_names['Freq_Label']
target2 = np.array(target2, dtype = int)
label_names = {1:'Buff1', 2:'Buff2', 3:'Buff3'}


data = data_nopeaks_names[data_nopeaks_names.columns[:-3]]
data = data.values
data_nopeaks_names
```

```python
# Names of files

nombres = []
for names in nopeaks:
    nombre = names['archivo']
    nombres.append(nombre)
```

### SOM OF FREQUENCIES

```python
# U-Matrix with raw data (not normalized)

som = MiniSom(16, 16, data.shape[1], sigma=4, learning_rate=.03)

# initialize the weights with random values
som.random_weights_init(data)

# train the SOM for 100 epochs
som.train_random(data, 200)

# create the U-matrix
u_matrix = som.distance_map()

ww = som.get_weights()
```

```python
# plot the U-matrix
import matplotlib.pyplot as plt
import matplotlib.lines as mlines

plt.figure(figsize=(8, 8))

plt.pcolor(som.distance_map().T, cmap='bone_r')  # plotting the distance map as background
plt.colorbar()

# Plotting the response for each pattern in the dataset
# different markers for each label
markers = ['o', 's', 'D'] #, '^', '<']
wdata = []
for cnt, xx in enumerate(data):
    w = som.winner(xx)  # getting the winner
    wdata.append(w)
    
    # place a marker on the winning position for the sample xx
    plt.plot(w[0]+.5, w[1]+.5, markers[target1[cnt]-1], markerfacecolor='None',
             markeredgecolor='#00FF00', markersize=8, markeredgewidth=1.5)

# Define the labels and handles for the legend
labels = ['BL = 384', 'BL = 672', 'BL = 960']#, '6400 Hz', '12000 Hz']
handles = [mlines.Line2D([], [], color='#00FF00', marker=m, linestyle='None', markersize=8, markeredgewidth=2) for m in markers]

# Add the legend to the plot
plt.legend(handles=handles, labels=labels, loc='upper center', bbox_to_anchor=(0.55, -0.03), ncol=5)

plt.savefig('Data Pre Processing/Graphs/U-Matrix-Frequencies.png', bbox_inches = 'tight', dpi=300)
plt.show()
```

```python
import matplotlib.gridspec as gridspec

Freq_label_names = {1:'700Hz', 2:'2000Hz', 3:'3300Hz', 4:'6400Hz', 5:'12000Hz'}
n_neurons = 16
m_neurons = 16
labels_map = som.labels_map(data, [Freq_label_names[t] for t in target2])

fig = plt.figure(figsize=(6, 6))
the_grid = gridspec.GridSpec(n_neurons, m_neurons, fig)
for position in labels_map.keys():
    label_fracs = [labels_map[position][l] for l in Freq_label_names.values()]
    plt.subplot(the_grid[n_neurons-1-position[1],
                         position[0]], aspect=1)
    patches, texts = plt.pie(label_fracs)

plt.title('Proportion of samples per Freq')
plt.legend(patches, Freq_label_names.values(), bbox_to_anchor=(13, 10), ncol=1)
#plt.savefig('Data Pre Processing/Graphs/Proportion of samples per Freq.png', bbox_inches = 'tight')
plt.show()
```

```python
som = MiniSom(16, 16, data.shape[1], sigma=0.5, learning_rate=.05,
              neighborhood_function='gaussian', random_seed=10)

#som = MiniSom(16, 16, data.shape[1], sigma=1.0, learning_rate=0.5)

max_iter = 200
q_error = []
t_error = []

for i in range(max_iter):
    rand_i = np.random.randint(len(data))
    som.update(data[rand_i], som.winner(data[rand_i]), i, max_iter)
    q_error.append(som.quantization_error(data))
    t_error.append(som.topographic_error(data))

plt.plot(np.arange(max_iter), q_error, label='quantization error')
plt.plot(np.arange(max_iter), t_error, label='topographic error')
plt.ylabel('error')
plt.xlabel('iteration index')
plt.legend()
plt.savefig('Data Pre Processing/Graphs/Convergence - Frequency.png', bbox_inches = 'tight')
plt.show()
```
